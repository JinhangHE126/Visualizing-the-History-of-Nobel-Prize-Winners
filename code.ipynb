````python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
````

````python
# Load data
nobel = pd.read_csv('datasets/nobel.csv')
nobel.head()
````

What is the most commonly awarded gender and birth country?

* Store your answers as string variables ```top_gender``` and ```top_country```.

````python
# top_gender = nobel['sex'].mode()[0]
top_gender = nobel['sex'].value_counts().index[0]
print(top_gender)

top_country = nobel['birth_country'].value_counts().index[0]
print(top_country)
print("\n The gender with the most Nobel laureates is :", top_gender)
print(" The most common birth country of Nobel laureates is :", top_country)
````

Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

* Store this as an integer called ```max_decade_usa```.

````python
nobel['decade'] = (nobel['year'] // 10) * 10
nobel['usa_born_winner'] = nobel['birth_country'] == 'United States of America'
prop_usa_winners = nobel.groupby('decade').agg(
    usa_born_winners = ('usa_born_winner', 'sum'),
    total_winners = ('usa_born_winner', 'count') 
)
ratio_usa_winners = prop_usa_winners['usa_born_winners'] / prop_usa_winners['total_winners']

max_decade_usa = ratio_usa_winners.idxmax()
print("\n The decade with the highest ratio of US-born Nobel Prize winners to total winners is :", max_decade_usa)
````

Which decade and Nobel Prize category combination had the highest proportion of female laureates?

* Store this as a dictionary called ```max_female_dict``` where the decade is the key and the category is the value. There should only be one key:value pair.

````python
nobel["female_winner"] = nobel["sex"] == "Female"
prop_female_winners = nobel.groupby(["decade", "category"], as_index=False).agg(
    female_winners=("female_winner", "sum"), total_winners=("female_winner", "count")
)


# calculate the ratio of female winners
prop_female_winners['ratio_female_laureates'] = prop_female_winners['female_winners'] / prop_female_winners['total_winners']

# print(prop_female_winners.head())

# find the decade and category with the highest proportion of female laureates
max_ratio_female_laureates = prop_female_winners[
    prop_female_winners["ratio_female_laureates"]
    == prop_female_winners["ratio_female_laureates"].max()
]
print(max_ratio_female_laureates)
# store in dictionary
max_female_dict = {
    int(max_ratio_female_laureates["decade"].values[0]): max_ratio_female_laureates["category"].values[0]
}
print(f"\n The decade and Nobel Prize category combination with the highest proportion of female laureates is:{max_female_dict}")
````

Who was the first woman to receive a Nobel Prize, and in what category?

* Save your string answers as ```first_woman_name``` and ```first_woman_category```.

````python
female_laureates = nobel[nobel['sex'] == 'Female']
female_laureates = female_laureates.sort_values('year')

first_woman_information = female_laureates.iloc[0]
# print(first_woman_information)

first_woman_name = first_woman_information['full_name']
first_woman_category = first_woman_information['category']
print(first_woman_name)
print(first_woman_category)

````

Which individuals or organizations have won more than one Nobel Prize throughout the years?

* Store the full names in a list named ```repeat_list```.

````python
counts = nobel["full_name"].value_counts()
repeats = counts[counts >= 2].index
repeat_list = list(repeats)

print("\n The repeat winners are :", repeat_list)
````
